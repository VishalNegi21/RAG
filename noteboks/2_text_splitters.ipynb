```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text = "Large documents are usually split into smaller chunks before generating embeddings. Chunking improves retrieval quality because the retriever can return only the relevant portions of a document instead of the entire file."
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=30 , chunk_overlap=9)
```

```python
chunks = splitter.split_text(text)
```

### Document Text Loader

```python
from langchain_community.document_loaders import TextLoader ,PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader(file_path="../data/text_data.txt")
texts = loader.load()
```

```python
pdf_loader = PyPDFLoader("../data/a.pdf")
pdfs = pdf_loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=150 , chunk_overlap=40)
```

```python
text_chunks = splitter.split_documents(texts)
len(text_chunks)
```

```python
pdf_chunks = splitter.split_documents(pdfs)
len(pdf_chunks)
```
